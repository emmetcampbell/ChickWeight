```python
import pandas as pd
import numpy as np
import scipy as sp

# visuals
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import ProbPlot

# modelling
from statsmodels.regression.linear_model import OLS
import statsmodels.api as sm
import statsmodels.formula.api as smf

color_scheme = {i: color for i, color in zip(range(1, 5), ["#D79922", "#C5CBE3", "#F13C20", "#4056A1"])}
ppt_ratio = (16, 9)
```

```python
chick_weight = pd.read_csv('C:\\Users\\emmet\\Desktop\\ChickWeight.csv')
# renaming column just for conformity
chick_weight = chick_weight.rename(columns={'weight': 'Weight'})
```

```python
chick_weight
```

```python
# pythons version of asfactor
chick_weight['Diet'] = chick_weight['Diet'].astype('category')
```

```python
'''
All observations should have a max Time of 21 anything with a max of less will be removed
'''
# which chick_id's have max Time of 42
mask = chick_weight.groupby('Chick')['Time'].max() == 21

# get id's of those with less
incompelete_observations = mask[mask == False].index

# drop incomplete observations
chick_weight = chick_weight.loc[~chick_weight.Chick.isin(incompelete_observations)]
```

```python
'''
Adding avg_daily_gain
'''
# weight change since previous observation
chick_weight['weight_change'] =  chick_weight.sort_values(by='Time').groupby(['Chick'])['Weight'].diff()

# days since previous observation
chick_weight['time_change'] = chick_weight.sort_values(by='Time').groupby(['Chick'])['Time'].diff()

# rate of weight change per day between observations
chick_weight['avg_daily_gain'] = chick_weight['weight_change'] / chick_weight['time_change']
```

```python
# creating a column for each category
X = pd.get_dummies(chick_weight[['Time', 'Diet']])
# adding constant for intercept
X = sm.add_constant(X) 
# dropped to make intercept interpretable
X = X.drop(columns='Diet_1')

y = chick_weight['Weight']

base_model = OLS(y, X).fit()
```

```python
base_model.summary2()
```

```python
f, ax = plt.subplots(figsize=ppt_ratio)

# yhat
predicted = base_model.fittedvalues
# e
residuals = y - predicted

sns.scatterplot(x = predicted,
                y = residuals,
                ax = ax)

ax.axhline(0, color='red')
ax.set_xlabel('Predicted')
ax.set_ylabel('Residuals')
ax.set_title('predicted vs residuals')
```

```python
f, ax = plt.subplots(figsize=ppt_ratio)
_ = sp.stats.probplot(residuals, plot=ax, fit=True)
```

Thoughts
We shouldn't include time 0 as it cannot be predicted (i.e. Diet 1 Time 0 gives no information)
Time 0 could however be an added feature 
Residuals aren't normal and appear to take a log shape

```python
# time 0 values for each chick
time_0 = chick_weight.sort_values(by = 'Time').groupby('Chick')['Weight'].first()
# adding time 0 values as column
chick_weight['start_weight'] = chick_weight['Chick'].map(time_0)

# removing all time 0 rows
data = chick_weight[chick_weight.Time != 0]

X = data[['Time', 'Diet', 'start_weight']]
# creating a column for each category
X = pd.get_dummies(X)
# adding intercept
X = sm.add_constant(X)
# make intercept interpretable
X = X.drop(columns = 'Diet_1')

y = np.log(data['Weight'])
```

```python
final_model = OLS(y, X).fit()
```

```python
final_model.summary2()
```

```python
f, ax = plt.subplots(figsize=ppt_ratio)

# yhat
predicted = final_model.fittedvalues
# e
residuals = y - predicted

sns.scatterplot(x = predicted,
                y = residuals,
                ax = ax)

ax.axhline(0, color='red')
ax.set_xlabel('Predicted')
ax.set_ylabel('Residuals')
ax.set_title('predicted vs residuals')
```

```python
# fitted values (need a constant term for intercept)
model_fitted_y = chick_weightlm8.fittedvalues

# model residuals
model_residuals = chick_weightlm8.resid

# normalized residuals
model_norm_residuals = chick_weightlm8.get_influence().resid_studentized_internal

# absolute squared normalized residuals
model_norm_residuals_abs_sqrt = np.sqrt(np.abs(model_norm_residuals))

# absolute residuals
model_abs_resid = np.abs(model_residuals)

# leverage, from statsmodels internals
model_leverage = chick_weightlm8.get_influence().hat_matrix_diag

# cook's distance, from statsmodels internals
model_cooks = chick_weightlm8.get_influence().cooks_distance[0]

#Enter Model Being Used
model_lm = final_model
```

```python
plot_lm_1 = plt.figure(1)
plot_lm_1.set_figheight(8)
plot_lm_1.set_figwidth(12)

plot_lm_1.axes[0] = sns.residplot(model_fitted_y, 'Weight', data=chick_weight,
                                  lowess=True,
                                  scatter_kws={'alpha': 0.5},
                                  line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plot_lm_1.axes[0].set_title('Residuals vs Fitted')
plot_lm_1.axes[0].set_xlabel('Fitted values')
plot_lm_1.axes[0].set_ylabel('Residuals')


# annotations
abs_resid = model_abs_resid.sort_values(ascending=False)
abs_resid_top_3 = abs_resid[:3]

for i in abs_resid_top_3.index:
    plot_lm_1.axes[0].annotate(i, 
                               xy=(model_fitted_y[i], 
                                   model_residuals[i]));
    
plot_lm_1.savefig('residualsvsfitted.png')
```

```python
QQ = ProbPlot(model_norm_residuals)
plot_lm_2 = QQ.qqplot(line='45', alpha=0.5, color='#4C72B0', lw=1)

plot_lm_2.set_figheight(8)
plot_lm_2.set_figwidth(12)

plot_lm_2.axes[0].set_title('Normal Q-Q')
plot_lm_2.axes[0].set_xlabel('Theoretical Quantiles')
plot_lm_2.axes[0].set_ylabel('Standardized Residuals');

# annotations
abs_norm_resid = np.flip(np.argsort(np.abs(model_norm_residuals)), 0)
abs_norm_resid_top_3 = abs_norm_resid[:3]

for r, i in enumerate(abs_norm_resid_top_3):
    plot_lm_2.axes[0].annotate(i, 
                               xy=(np.flip(QQ.theoretical_quantiles, 0)[r],
                                   model_norm_residuals[i]));

    
plot_lm_2.savefig('normalQQ.png')
```

```python
plot_lm_3 = plt.figure(3)
plot_lm_3.set_figheight(8)
plot_lm_3.set_figwidth(12)

plt.scatter(model_fitted_y, model_norm_residuals_abs_sqrt, alpha=0.5)
sns.regplot(model_fitted_y, model_norm_residuals_abs_sqrt, 
            scatter=False, 
            ci=False, 
            lowess=True,
            line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plot_lm_3.axes[0].set_title('Scale-Location')
plot_lm_3.axes[0].set_xlabel('Fitted values')
plot_lm_3.axes[0].set_ylabel('$\sqrt{|Standardized Residuals|}$');

# annotations
abs_sq_norm_resid = np.flip(np.argsort(model_norm_residuals_abs_sqrt), 0)
abs_sq_norm_resid_top_3 = abs_sq_norm_resid[:3]

for i in abs_norm_resid_top_3:
    plot_lm_3.axes[0].annotate(i, 
                               xy=(model_fitted_y[i], 
                                   model_norm_residuals_abs_sqrt[i]));
    
plot_lm_3.savefig('scalelocation.png')
```

```python
plot_lm_4 = plt.figure(4)
plot_lm_4.set_figheight(8)
plot_lm_4.set_figwidth(12)

plt.scatter(model_leverage, model_norm_residuals, alpha=0.5)
sns.regplot(model_leverage, model_norm_residuals, 
            scatter=False, 
            ci=False, 
            lowess=True,
            line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plot_lm_4.axes[0].set_title('Residuals vs Leverage')
plot_lm_4.axes[0].set_xlabel('Leverage')
plot_lm_4.axes[0].set_ylabel('Standardized Residuals')

# annotations
leverage_top_3 = np.flip(np.argsort(model_cooks), 0)[:3]

for i in leverage_top_3:
    plot_lm_4.axes[0].annotate(i, 
                               xy=(model_leverage[i], 
                                   model_norm_residuals[i]))
    
# shenanigans for cook's distance contours
def graph(formula, x_range, label=None):
    x = x_range
    y = formula(x)
    plt.plot(x, y, label=label, lw=1, ls='--', color='red')

p = len(model_lm.params) # number of model parameters

graph(lambda x: np.sqrt((0.5 * p * (1 - x)) / x), 
      np.linspace(0.001, 0.025, 50), 
      'Cook\'s distance') # 0.5 line

graph(lambda x: np.sqrt((1 * p * (1 - x)) / x), 
      np.linspace(0.001, 0.025, 50)) # 1 line

plt.legend(loc='upper right');

    
plot_lm_4.savefig('residualsvsleverage.png')
```
